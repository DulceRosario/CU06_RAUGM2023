# Caso de uso 3: Kernel Matlab


## Climate Data Visualization and Analysis Toolbox 

Esta toolbox contiene un conjunto estándar de funciones de MATLAB para analizar y visualizar datos climáticos.

### Carga de datos

La función *imread*  lee una imagen desde el archivo. En este caso, la variable **I** contiene la información de la imagen mientras que la variable *cmap* contiene la paleta de colores asociada al mapa.

```matlab
dataFile = "N_202109_extent_v3.0.tif";
[I,cmap] = imread(dataFile);
imshow(I,cmap)
title("Extensión del hielo marino del Ártico en septiembre de 2021")
```

La función *readtable* nos permite leer una tabla de datos de un archivo separados por comas csv. En este caso, el archivo contiene las anomalias de temperatura Global.  

```matlab

dataGbl = readtable("GlobalTempAnomaly.csv");
```

Para acceder a los datos en las columnas se utiliza el operador . 

```matlab

yearGbl = dataGbl.Year; 
anomalyGbl = dataGbl.Value; 

```

 ## Generación de gráfica 

```matlab
plot(yearGbl,anomalyGbl) 
title("Annual Global Temperature Anomaly")
ylabel("Anomalia (°C)")
xlabel("Año")
```

Ahora leemos las anomalías de temperatura en Norte America

```matlab

dataNA = readtable("NATempAnomaly.csv");
```

Se extraen los datos para Norteamérica

```matlab
yearNA = dataNA.Year; 
anomalyNA = dataNA.Value;
```

Se crea una gráfica comparativa de los datos. 
- Primero se genera la gŕafica Global 
- *hold on* en los mismos ejes
- Se grafican las anomalias regionales
- A partir *hold off* la siguiente instrucción *plot* se generá por separado 
- Se pone título y leyendas

```matlab
plot(yearGbl,anomalyGbl)
hold on
plot(yearNA,anomalyNA) 
hold off 
title("Anomalia de temperatura anual ")
ylabel("Anomalía (°C)")
xlabel("Year")
legend("Global","Norteamerica","Location","northwest")
```

Se calcula la media móvil  y se gŕafica

```matlab
mmAnomalyGbl = movmean(anomalyGbl,10);
% Create the plot
plot(yearGbl,mmAnomalyGbl) 
title("Anomalía de temperatura (media móvil 10-años)")
ylabel("Anomaly (°C)")
xlabel("Year")
```

## Mapping Toolbox™

Proporciona algoritmos y funciones para transformar datos geográficos y crear visualizaciones de mapas. Puede visualizar sus datos en un contexto geográfico, crear visualizaciones de mapas a partir de más de 60 proyecciones de mapas y transformar datos de una variedad de fuentes en un sistema de coordenadas geográficas uniforme.

```matlab
eox = wmsfind("eox");
```

```matlab

eox_marble = refine(eox,"blue marble");
```

```matlab
eox_marble_limits = refineLimits(eox_marble,"Latlim",[-90 90]);
```

```matlab
eox_update = wmsupdate(eox_marble_limits);
```

```matlab
figure
worldmap europe
mstruct = gcm;
latlim = mstruct.maplatlimit;
lonlim = mstruct.maplonlimit;

[A,R] = wmsread(eox_update,"Latlim",latlim,"Lonlim",lonlim);


geoshow(A,R)
title(eox_update.LayerTitle)
plabel off
mlabel off
```
